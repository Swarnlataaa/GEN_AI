```python
import tensorflow as tf

# Load the saved generator model
generator = tf.keras.models.load_model('gan_generator_model.h5')

# Generate new images using the loaded generator
noise = tf.random.normal([10, 100])  # Example: Generate 10 random images
generated_images = generator.predict(noise)

# Display or save the generated images (optional)
import matplotlib.pyplot as plt

for i in range(10):  # Assuming you generated 10 images
    plt.imshow(generated_images[i].reshape(28, 28), cmap='gray')
    plt.axis('off')
    plt.show()
```
